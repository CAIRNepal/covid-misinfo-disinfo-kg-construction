```python
from pyshacl import validate
from rdflib import Graph

# Load data and shapes graphs
data_graph = Graph().parse("MisinfoGraphInstances.ttl", format="turtle")
shapes_graph = Graph().parse("covid_misinfo_shapes.ttl", format="turtle")

# Validate
conforms, results_graph, results_text = validate(
    data_graph,
    shacl_graph=shapes_graph,
    inference="none", 
    debug=True
)

# Print results
print(f"Conforms: {conforms}")
print(results_text)

# Serialize validation report
results_graph.serialize("validation_report.ttl", format="turtle")
```
